## Selecting a single column by column name
- simpliest data selection case

syntax:
```
df['col_name']
```
Note this is the same syntax for a dictionary. Remember we can think of a pd.DataFrame as a dictionary where the keys are the col. names. 

Ex: what lvl has the sea risen?

````python
annual_rise = df['annual_sea_level_rise']

print(type(annual_rise))

annual_rise
````

`df['col_name']` is an example of selecting **by label**
**selecting by label** means we can select data from the data frame using the names of the columns, not their position

## Selecting a single column w/ attribute syntax

syntx:
```
df.col_name
```
Ex:

````python
# select annual sea lvl rise

df.annual_sea_level_rise
````

## Selecting multiple columns using a list of col names
list syntax [ , , ]:

selecting list syntax:
```
df[['col1', 'col2', 'col3']]
```
contains the names of the columns we wish to select, inside [] for lists


````python
europe_asia = df[['europe', 'asia']]    # creates a dataframe when more than one col. is called. series when only one is called

print(type(europe_asia))

print(europe_asia.shape)               # expect 43 rows and 2 columns

europe_asia                            # confirmed
````

## Selecting multiple columns using a slice
`loc` used for selction
syntax:
```
df.loc[row-selection , column_selection]
```
where
- `row-selection` = rows wanted to subset from df
- `column-selection` = cols wanted to subset from df

````python
# Slect all columns between europe and antarctica 

df.loc[ : , 'europe':'antarctica'].head()
````

Notes
- `row-selection` = ':' colon = select all the rows
- `column-selection` = slice `'europe':'antarctica'`
- slicing in panda is inclusive - different than base python and numpy, where the end point is *not* included

# Selecting Rows

## Selecting rows using a condition
syntax:
```
df[ condition_on_rows]
```

Ex: what's data like after 1996?

````python
# Select all rows w/ data > 1996

df[ df['year']>1996]
````

### Breaking it down
- in the `df[condition_on_rows]` syntax, we have that 'condition_on_rows' = `df['year']>1996` 
- checks which rows have values > 1996 for the year column
- `df['year']>1996 is a panda series with boolean values (T/F) are often called **masked**
- when we pass a `pd.Series` with boolean values (mask) to the selectedion brackets [], we keep only those rows with a True value in the pd.Series

Example: select data from years 1970 - 1979 (including 79)

````python
# Ends are snipped when using range

df[ df.year.isin(range(1970,1980)) ]
````

````python
df.year.isin(range(1970,1980))
````

## Selecting rows with multiple conditions

we can combine multiple conditions by surrounding each one in parenthesis `()` and using the **or operator** `|` and the **and operator** `&` 

syntax:
```
# select rows of df that satisfy cond. 1 OR 2

df[ (condition1) | (condition2)]

and

# select rows that satisfy both conditions

df[ (condition1) & (condition2)]
```

Ex:

````python
# select rows w/ annual sea lvl rise of >0.8 mm

df [ (df['annual_sea_level_rise']<0) | (df['annual_sea_level_rise']>0.8)]
````

## Selecting rows by position

Sometimes we want to select rows on actual position for this

use `iloc` selection, syntax is:

```
df.iloc[row-indicies]
```

`iloc` = **integer** location based on indexing

````python
# select the fifth row = index 4

df.iloc[4]
````

````python
# select rows 23 through 30, including 30

df.iloc[23:31]
````

## Selecting rows and columns simultaneously

can do this using `loc` (for labels or conditons) or `iloc` (integer positon)

## Selecting rows and columns by label or conditions
`loc`
```
df.loc[ row-selection, column-selection]
```

`row-selection` & `column-selection` can be a condition or subset of lables from the index of the column names

Ex:

````python
# select change in glacial volume in Europe per year after 2000

# columns: change in glacial volume, europe
# rows: years after 2000

df.loc[ df['year']>2000, ['year', 'europe']]
````

## Selecting rows and columns by position

We use `iloc` with snytax:
```
df.iloc[ row-indices, column-indices]
```
Ex:

````python
# selecting rows with indicies 3-7 (including 7) and the 4th and 5th column

df.iloc[3:8, [3, 4] ]
````
